```python
import sys
sys.path.append("..")
import numpy as np
import matplotlib.pyplot as plt

import lauetools.LaueTools.IOimagefile as IOimage


#x1, x2, y1, y2 =50,55,40,45
x1, x2, y1, y2 = 56,61,37,42
x1, x2, y1, y2 = 0,5,8,13

fig = plt.figure(constrained_layout=False,figsize=(12,12))
gs = fig.add_gridspec(nrows=5, ncols=5, left=0, right=.5,bottom=0,top=.5, wspace=0.002,hspace=0.02)

#x1, x2, y1, y2 = 0,5,0,5

#s_range=range(64)
#c_range=range(146)

s_range=range(y1,y2)
c_range=range(x1,x2)

for i in s_range:
    for j in c_range:
        if (i >= 100):
            str_i=str(i)
        else:
            if (i >= 10):
                str_i='0'+str(i)
            else:
                str_i='00'+str(i)
        if (j >= 100):
            str_j=str(j)
        else:
            if (j >= 10):
                str_j='0'+str(j)
            else:
                str_j='00'+str(j)
        path="../XRD_images/2nd_saving/"+"140_S"+str_i+"_C"+str_j+".TIFF"
        im=IOimage.readCCDimage(path,CCDLabel='TIFF Format')[0]
        ax = fig.add_subplot(gs[i-min(s_range),j-min(c_range)])
        ax.axis('off')
        ax.text(0,0,"S"+str_i+"_C"+str_j,size='large',verticalalignment='top',color='red')
        im=ax.imshow(im,cmap='gray')
```
