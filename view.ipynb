```python
import matplotlib.pyplot as plt
import numpy as np
load=140

if load==100:
    pre='100'
    dirname="../XRD_images/1st_saving/"
elif load==140:
    pre='140'
    dirname="../XRD_images/2nd_saving/"
si, ei, sj, ej = 0,8,0,16

#count=1
#fig,ax =  plt.subplots(ei-si,ej-sj)

ims=np.zeros(((ei-si)*1024,(ej-sj)*1024))

for i in range(si,ei):
    for j in range(sj,ej):
        if (i >= 100):
            str_i=str(i)
        else:
            if (i >= 10):
                str_i='0'+str(i)
            else:
                str_i='00'+str(i)
        if (j >= 100):
            str_j=str(j)
        else:
            if (j >= 10):
                str_j='0'+str(j)
            else:
                str_j='00'+str(j)
        x=IOimage.readCCDimage(filename=pre+'_S'+str_i+'_C'+str_j+".TIFF",CCDLabel='TIFF Format',dirname=dirname)
        ims[i*1024:(i+1)*1024,j*1024:(j+1)*1024]=x[0]
        #ax.plot(i-si,j-sj)
        #ax[i-si,j-sj].imshow(im)
        #count += 1
#plt.set_xaxis('off')
#plt.yaxis('off')
plt.axis('off')
plt.imshow(np.log(ims),cmap='gray')

plt.savefig("ims.png",format='png',bbox_inches ='tight')
```
