```python
import numpy as np
import matplotlib.pyplot as plt

x=np.loadtxt('curve.txt',skiprows=3)

plt.figure(figsize=(15,15))

plt.plot(np.abs(x[:,0]),x[:,1],'k')

plt.xlabel("Elongation (%)",size='xx-large')
plt.ylabel("Stress (MPa)",size='xx-large')
plt.xticks(size='xx-large')
plt.yticks(size='xx-large')


plt.savefig('curve.tif',format='TIFF',bbox_inches ='tight')
```
